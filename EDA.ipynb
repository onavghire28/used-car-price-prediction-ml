```python
data['Mileage_kmpl'] = data['Mileage_kmpl'].fillna(data['Mileage_kmpl'].mean())
data['Engine_CC'] = data['Engine_CC'].fillna(data['Engine_CC'].mean())
data['Horsepower'] = data['Horsepower'].fillna(data['Horsepower'].mean())
```

```python
data.isnull().sum()
```

```python
data['Fuel_Type'] = data['Fuel_Type'].fillna(data['Fuel_Type'].mode()[0])
data['Transmission'] = data['Transmission'].fillna(data['Transmission'].mode()[0])
data['City'] = data['City'].fillna(data['City'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data = data.drop('Color', axis = 1)
```
