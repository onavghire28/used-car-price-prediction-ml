```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Output.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.drop('Registration_Age', axis = 1)	
```

```python
data.shape
```

```python
data['Car_Age'] = 2026 - data['Year']
```

```python
data.head()
```

```python
data['Transmission'] = data['Transmission'].map({'Automatic': 0, 'Manual': 1})
```

```python
data.head()
```

```python
data = data.drop(columns=['Horsepower','City','Service_History','Accidents','Tax_Paid','Number_of_Doors'], axis = 1)
```

```python
data.shape
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns = ['Brand','Model','Fuel_Type'], dtype = int)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Owner_Type'] = le.fit_transform(data['Owner_Type'])
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler_feature = ['Year', 'Mileage_kmpl', 'Engine_CC','Owner_Type', 'Kms_Driven','Car_Age']

data[scaler_feature] = scaler.fit_transform(data[scaler_feature])
```

```python
data.head()
```

```python
X = data.drop('Price', axis = 1)
```

```python
X.shape
```

```python
y = data['Price']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
xgb = XGBRegressor()
```

```python
scores = cross_val_score(
    xgb,
    X,
    y,
    cv=5,
    scoring='r2'
)
```

```python
print("CV Scores:", scores)
print("Average CV R2:", scores.mean())
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

random = RandomizedSearchCV(
    estimator= xgb,
    param_distributions=params,
    n_iter = 10,
    cv = 5,
    scoring = 'r2',
    random_state = 42

)
```

```python
random.fit(X_train, y_train)
```

```python
print(random.best_params_)
```

```python
print(random.best_score_)
```

```python
best_model = random.best_estimator_
print(best_model)
```

```python
y_pred = random.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_random = r2_score(y_test, y_pred)
mae_random = mean_absolute_error(y_test, y_pred)
mse_random = mean_squared_error(y_test, y_pred)
```

```python
print("R2 Score: ", r2_random)
print("MSE: ", mse_random)
print("MAE: ", mae_random)
```

```python
import pickle

# Save model
with open('model.pkl', 'wb') as file:
    pickle.dump(random, file)
```
